```python
# Question: Do per-game sale of video games go up or down over the life of a console
data = pd.read_csv('../data/vgsales.csv')
print(data.columns)

# Print off the top platforms
print(data['Platform'].value_counts().head(8))
```

```python
# Filter
dataDS = data[ data['Platform'] == "DS" ]
dataPS2 = data[ data['Platform'] == "PS2" ]
dataPS3 = data[ data['Platform'] == "PS3" ]
dataWii = data[ data['Platform'] == "Wii" ]
dataX360 = data[ data['Platform'] == "X360" ]

# Clean: Owner_Type
for idx, row in dataDS.iterrows():
    if row['Year'] < 2004:
        dataDS.at[idx, 'Year'] = 2004
```

```python
# Graph setup
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid(True)

# Graphs
plt.title('DS')
plt.scatter(dataDS['Year'], dataDS['Global_Sales'], alpha=0.7)
plt.show()

plt.title('PS2')
plt.scatter(dataPS2['Year'], dataPS2['Global_Sales'], alpha=0.7)
plt.show()

plt.title('PS3')
plt.scatter(dataPS3['Year'], dataPS3['Global_Sales'], alpha=0.7)
plt.show()

plt.title('Wii')
plt.scatter(dataWii['Year'], dataWii['Global_Sales'], alpha=0.7)
plt.show()

plt.title('X360')
plt.scatter(dataX360['Year'], dataX360['Global_Sales'], alpha=0.7)
plt.show()
```
