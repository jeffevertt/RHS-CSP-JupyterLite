```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data (used car database from the UK)
data = pd.read_csv('../data/usedCarsUK-simple.csv')
```

---
### Inspect the data
- Each row is a data item which specifies values for all data-fields
- The columns are the data fields (type of data each row has)

```python
# Display all the columns & the first 5 rows
display( data.columns )
display( data.values[:5] )
```

---
Create a few graphs to visualize the data

```python
# Bar graph of mileage
data['Kilometers_Driven'].plot(kind='bar')
plt.title('Used Cars (mileage, km)')
plt.show()
```

```python
# Histogram (mileage)
data['Kilometers_Driven'].hist(bins=10)
plt.title('Used Cars (mileage, km)')
plt.show()
```

```python
# Histogram (transmission)
data['Transmission'].hist(bins=10)
plt.title('Used Cars (transmission)')
plt.show()
```

```python
# Cleaning (tranmission type)
for idx, row in data.iterrows():
    # Access row data using row['column_name']
    if row['Transmission'] == 'manual':
        data.at[idx, 'Transmission'] = 'Manual'

# Quick inspection (the first row had one a car with lower case 'manual')
display( data.values[:3] )
```
